```python
o365 = open('/local1/bryanzhou008/jarvis/project_jarvis/bryan/ood_categories/o365_categories.txt', 'r')
ego4d = open('/local1/bryanzhou008/jarvis/project_jarvis/bryan/ood_categories/scod_label_categories.txt', 'r')

o365 = o365.readlines()
ego4d = ego4d.readlines()

o365_clean = [obj.strip() for obj in o365]
o365_clean.sort()
print(o365_clean)
```

```python
ego_clean = [o.strip().replace("_", " ").split("(", 1)[0] for o in ego4d]
print(ego_clean)
```

```python
exact_match = [(obj in o365_clean) for obj in ego_clean]

print(sum(exact_match))
print(len(exact_match))
print(sum(exact_match)/len(exact_match))
```

```python
sub_match = []

for obj in ego_clean:
    found = False
    for t in o365_clean:
        if obj in t:
            found = True
    sub_match.append(found)

print(sum(sub_match))
print(len(sub_match))
print(sum(sub_match)/len(sub_match))
```
