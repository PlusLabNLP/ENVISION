# Dataset Annotation Visualization

```python
import base64
from IPython.display import Image, display, HTML
import random
from momaapi import MOMA, AnnVisualizer
```

Initialize MOMA API.

```python
dir_moma = "/local1/telinwu/research/resources/MOMA-LRG"
moma = MOMA(dir_moma)
```

```python
visualizer = AnnVisualizer(moma)
```

Visualize a random higher-order interaction that contains the object camera.

```python
ids_hoi = moma.get_ids_hoi(
    cnames_object=[
        "table"
    ]
)
id_hoi = random.choice(ids_hoi)
print(id_hoi)

path = visualizer.show_hoi(id_hoi, vstack=False)

display(Image(path))
```

Visualize a random sub-activity.

```python
ids_sact = moma.get_ids_sact()
id_sact = random.choice(ids_sact)
path = visualizer.show_sact(id_sact, vstack=False)

with open(path, "rb") as f:
    b64 = base64.b64encode(f.read()).decode("ascii")
HTML(f"<img src='data:image/gif;base64,{b64}' width='100%' />")
```
